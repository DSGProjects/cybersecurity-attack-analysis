```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine("sqlite:///../data/cybersecurity.db")
```

```python
df = pd.read_csv("../data/cybersecurity_attacks.csv")
```

```python
engine = create_engine("sqlite:///../data/cybersecurity.db")
```

```python
df.columns = (
    df.columns
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("-", "_")
        .str.replace("/", "_")
)

```

```python
# Insertar datos
df.to_sql(
    "cybersecurity_attacks",
    con=engine,
    if_exists="replace",  
    index=False
)

print("Datos insertados correctamente")
```
